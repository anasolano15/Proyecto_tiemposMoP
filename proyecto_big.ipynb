```python
df = pd.read_excel('/Big Data (1) (1).xlsx')

# Calcula el promedio de la columna 'exportacion'
promedio = df['Engine_rpm'].mean()
print("El promedio de la exportación es:", promedio)

moda_columna = df['Engine_rpm'].mode()
print("la moda es:", moda_columna)

desviacion_estandar_columna = df['Engine_rpm'].std()
print("La desviación estándar de la columna es", desviacion_estandar_columna)

varianza_muestra_columna = df['Engine_rpm'].var(ddof=1)
print("La varianza de la muestra de la columna es", varianza_muestra_columna)

curtosis_columna = df['Engine_rpm'].kurtosis()
print("La curtosis de la columna es", curtosis_columna)

coef_asimetria_columna = df['Engine_rpm'].skew()
print("El coeficiente de asimetría de la columna es:", coef_asimetria_columna)

rango_columna = df['Engine_rpm'].max() - df['Engine_rpm'].min()
print("El rango de la columna es:", rango_columna)

minimo_columna = df['Engine_rpm'].min()
print("El mínimo de la columna es:", minimo_columna)

maximo_columna = df['Engine_rpm'].max()
print("El máximo de la columna es:", maximo_columna)

suma_columna = df['Engine_rpm'].sum()
print("La suma de la columna es:", suma_columna)

conteo_columna = df['Engine_rpm'].count()
print("El número de valores en la columna es:", conteo_columna)






```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Leer el archivo Excel
df = pd.read_excel('/Big Data (1) (1).xlsx')

# Calcular los estadísticos
promedio = df['Fuel_pressure'].mean()
moda_columna = df['Fuel_pressure'].mode()[0]
desviacion_estandar_columna = df['Fuel_pressure'].std()
varianza_muestra_columna = df['Fuel_pressure'].var(ddof=1)
curtosis_columna = df['Fuel_pressure'].kurtosis()
coef_asimetria_columna = df['Fuel_pressure'].skew()
rango_columna = df['Fuel_pressure'].max() - df['Fuel_pressure'].min()
minimo_columna = df['Fuel_pressure'].min()
maximo_columna = df['Fuel_pressure'].max()
suma_columna = df['Fuel_pressure'].sum()
conteo_columna = df['Fuel_pressure'].count()

# Crear DataFrame con los estadísticos
data = {
    'Estadístico': ['Promedio', 'Moda', 'Desviación Estándar', 'Varianza',
                    'Curtosis', 'Coeficiente de Asimetría', 'Rango',
                    'Mínimo', 'Máximo', 'Suma', 'Conteo'],
    'Valor': [promedio, moda_columna, desviacion_estandar_columna, varianza_muestra_columna,
              curtosis_columna, coef_asimetria_columna, rango_columna,
              minimo_columna, maximo_columna, suma_columna, conteo_columna]
}
df_estadisticos = pd.DataFrame(data)

# Graficar el diagrama de barras
plt.figure(figsize=(10, 6))
plt.bar(df_estadisticos['Estadístico'], df_estadisticos['Valor'], color='skyblue')

# Añadir título y etiquetas de ejes
plt.title('Estadísticas de la columna "Fuel_pressure"')
plt.xlabel('Estadístico')
plt.ylabel('Valor')

# Rotar las etiquetas del eje x para mejorar la legibilidad
plt.xticks(rotation=45, ha='right')

# Mostrar la gráfica
plt.tight_layout()
plt.show()
```
